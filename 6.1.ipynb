```python
import pandas as pd
keywords = pd.read_csv('C:/Users/danii/OneDrive/Презентации аналитика/python/Python_6_pandas.ipynb/keywords.csv')
```

```python
geo_data = {

    'Центр': ['москва', 'тула', 'ярославль'],

    'Северо-Запад': ['петербург', 'псков', 'мурманск'],

    'Дальний Восток': ['владивосток', 'сахалин', 'хабаровск']

}
geo_data1 = geo_data['Центр']
geo_data2 = geo_data['Северо-Запад']
geo_data3 = geo_data['Дальний Восток']

geo_data = pd.Series(geo_data)

type(geo_data)
geo_data
```

```python
def region (row):
        if row['keyword'] in geo_data['Центр']:
            return 'Центр'
        if row['keyword'] in geo_data['Северо-Запад']:
            return 'Северо-Запад'
        if row['keyword'] in geo_data['Дальний Восток']:
            return 'Дальний Восток'
        else:
            return 'undefined'
```

```python
keywords['regions'] = keywords.apply(region, axis = 1)
```
